```python
import os
import time
from dataclasses import dataclass
from typing import Any, Dict, Optional, Callable, Tuple

import torch
from torch.utils.data import DataLoader


# -----------------------------
# Helpers
# -----------------------------
def _to_device(x: Any, device: torch.device):
    if torch.is_tensor(x):
        return x.to(device, non_blocking=True)
    if isinstance(x, dict):
        return {k: _to_device(v, device) for k, v in x.items()}
    if isinstance(x, (list, tuple)):
        return type(x)(_to_device(v, device) for v in x)
    return x


def _default_step_fn(
    model: torch.nn.Module,
    batch: Any,
    loss_fn: Callable,
) -> Tuple[torch.Tensor, Dict[str, float]]:
    """
    Default "step" assumes:
      - model(**batch) OR model(batch) returns logits or a dict with "logits"
      - batch contains labels in batch["labels"] (or second tuple element)
    You almost certainly will customize this for HuBERT.
    """
    if isinstance(batch, dict):
        outputs = model(**{k: v for k, v in batch.items() if k != "labels"})
        labels = batch.get("labels", None)
    elif isinstance(batch, (list, tuple)) and len(batch) == 2:
        inputs, labels = batch
        outputs = model(inputs)
    else:
        raise ValueError("Batch format not supported by default step_fn. Provide a custom step_fn.")

    # Handle outputs that might be dict-like
    if isinstance(outputs, dict):
        logits = outputs.get("logits", None)
        if logits is None:
            raise ValueError("Model output dict missing 'logits'. Provide custom step_fn.")
    else:
        logits = outputs

    if labels is None:
        raise ValueError("Missing labels in batch for default step_fn. Provide custom step_fn.")

    loss = loss_fn(logits, labels)

    # Example metric: accuracy for classification-like tasks
    with torch.no_grad():
        preds = torch.argmax(logits, dim=-1)
        acc = (preds == labels).float().mean().item()

    metrics = {"loss": loss.detach().item(), "acc": acc}
    return loss, metrics


def save_checkpoint(
    path: str,
    model: torch.nn.Module,
    optimizer: torch.optim.Optimizer,
    epoch: int,
    best_metric: Optional[float],
    scaler: Optional[torch.cuda.amp.GradScaler] = None,
    scheduler: Optional[Any] = None,
    extra: Optional[Dict[str, Any]] = None,
):
    ckpt = {
        "epoch": epoch,
        "model_state": model.state_dict(),
        "optimizer_state": optimizer.state_dict(),
        "best_metric": best_metric,
    }
    if scaler is not None:
        ckpt["scaler_state"] = scaler.state_dict()
    if scheduler is not None:
        ckpt["scheduler_state"] = scheduler.state_dict()
    if extra is not None:
        ckpt["extra"] = extra
    os.makedirs(os.path.dirname(path), exist_ok=True) if os.path.dirname(path) else None
    torch.save(ckpt, path)


def load_checkpoint(
    path: str,
    model: torch.nn.Module,
    optimizer: Optional[torch.optim.Optimizer] = None,
    scaler: Optional[torch.cuda.amp.GradScaler] = None,
    scheduler: Optional[Any] = None,
    map_location: str = "cpu",
) -> Dict[str, Any]:
    ckpt = torch.load(path, map_location=map_location)
    model.load_state_dict(ckpt["model_state"])
    if optimizer is not None and "optimizer_state" in ckpt:
        optimizer.load_state_dict(ckpt["optimizer_state"])
    if scaler is not None and "scaler_state" in ckpt:
        scaler.load_state_dict(ckpt["scaler_state"])
    if scheduler is not None and "scheduler_state" in ckpt:
        scheduler.load_state_dict(ckpt["scheduler_state"])
    return ckpt


@dataclass
class TrainConfig:
    device: str = "cuda" if torch.cuda.is_available() else "cpu"
    eval_every: int = 10               # run eval every N epochs (10 per your request)
    grad_accum_steps: int = 1
    max_grad_norm: Optional[float] = 1.0
    use_amp: bool = True
    log_every: int = 50                # steps
    ckpt_dir: str = "./checkpoints"
    ckpt_name: str = "model"
    monitor: str = "loss"              # metric key to monitor from eval
    mode: str = "min"                  # "min" or "max" for monitor metric
    resume_path: Optional[str] = None
    save_last: bool = True
    save_best: bool = True


# -----------------------------
# Core loops
# -----------------------------
@torch.no_grad()
def evaluate(
    model: torch.nn.Module,
    dataloader: DataLoader,
    loss_fn: Callable,
    device: torch.device,
    step_fn: Callable = _default_step_fn,
    max_batches: Optional[int] = None,
) -> Dict[str, float]:
    model.eval()

    sums: Dict[str, float] = {}
    n = 0

    for b_idx, batch in enumerate(dataloader):
        if max_batches is not None and b_idx >= max_batches:
            break

        batch = _to_device(batch, device)
        loss, metrics = step_fn(model, batch, loss_fn)

        # Force include loss in metrics if step_fn did not
        if "loss" not in metrics:
            metrics["loss"] = loss.item()

        for k, v in metrics.items():
            sums[k] = sums.get(k, 0.0) + float(v)
        n += 1

    if n == 0:
        return {"loss": float("nan")}

    return {k: v / n for k, v in sums.items()}


def train_eval_loop(
    model: torch.nn.Module,
    train_loader: DataLoader,
    num_epochs: int,
    loss_fn: Callable,
    optimizer: torch.optim.Optimizer,
    val_loader: Optional[DataLoader] = None,
    config: Optional[TrainConfig] = None,
    scheduler: Optional[Any] = None,
    step_fn: Callable = _default_step_fn,
    max_train_batches: Optional[int] = None,
    max_eval_batches: Optional[int] = None,
) -> Dict[str, Any]:
    """
    Generic train/eval loop with checkpointing.

    Args:
      model: torch.nn.Module
      train_loader: DataLoader
      num_epochs: int
      loss_fn: callable
      optimizer: torch optimizer
      val_loader: optional DataLoader
      config: TrainConfig
      scheduler: optional lr scheduler
      step_fn: function(model, batch, loss_fn) -> (loss_tensor, metrics_dict)
              You should override this for HuBERT to match your model/batch format.
      max_train_batches / max_eval_batches: for debugging

    Returns:
      dict with training history and best checkpoint info
    """
    if config is None:
        config = TrainConfig()

    device = torch.device(config.device)
    model.to(device)

    scaler = torch.cuda.amp.GradScaler(enabled=(config.use_amp and device.type == "cuda"))

    # Resume (optional)
    start_epoch = 0
    best_metric = None
    if config.resume_path is not None and os.path.isfile(config.resume_path):
        ckpt = load_checkpoint(
            config.resume_path,
            model=model,
            optimizer=optimizer,
            scaler=scaler,
            scheduler=scheduler,
            map_location="cpu",
        )
        start_epoch = int(ckpt.get("epoch", 0)) + 1
        best_metric = ckpt.get("best_metric", None)
        print(f"Resumed from {config.resume_path} at epoch {start_epoch} (best_metric={best_metric})")

    # Checkpoint paths
    ckpt_dir = config.ckpt_dir
    os.makedirs(ckpt_dir, exist_ok=True)
    last_path = os.path.join(ckpt_dir, f"{config.ckpt_name}_last.pt")
    best_path = os.path.join(ckpt_dir, f"{config.ckpt_name}_best.pt")

    history = {
        "train": [],
        "eval": [],
        "best_metric": best_metric,
        "best_epoch": None,
        "last_path": last_path,
        "best_path": best_path,
    }

    def is_better(new: float, best: Optional[float]) -> bool:
        if best is None:
            return True
        return (new < best) if config.mode == "min" else (new > best)

    global_step = 0

    for epoch in range(start_epoch, num_epochs):
        model.train()
        epoch_sums: Dict[str, float] = {}
        epoch_steps = 0

        t0 = time.time()
        optimizer.zero_grad(set_to_none=True)

        for b_idx, batch in enumerate(train_loader):
            if max_train_batches is not None and b_idx >= max_train_batches:
                break

            batch = _to_device(batch, device)

            with torch.cuda.amp.autocast(enabled=(scaler.is_enabled())):
                loss, metrics = step_fn(model, batch, loss_fn)

                # Normalize for grad accumulation
                loss_to_backprop = loss / max(1, config.grad_accum_steps)

            scaler.scale(loss_to_backprop).backward()

            # accumulate metrics
            if "loss" not in metrics:
                metrics["loss"] = loss.detach().item()
            for k, v in metrics.items():
                epoch_sums[k] = epoch_sums.get(k, 0.0) + float(v)
            epoch_steps += 1
            global_step += 1

            # update weights every grad_accum_steps
            if global_step % config.grad_accum_steps == 0:
                if config.max_grad_norm is not None:
                    scaler.unscale_(optimizer)
                    torch.nn.utils.clip_grad_norm_(model.parameters(), config.max_grad_norm)

                scaler.step(optimizer)
                scaler.update()
                optimizer.zero_grad(set_to_none=True)

                if scheduler is not None:
                    scheduler.step()

            # basic logging
            if (b_idx + 1) % config.log_every == 0:
                avg_loss = epoch_sums.get("loss", 0.0) / max(1, epoch_steps)
                lr = optimizer.param_groups[0]["lr"]
                print(
                    f"Epoch {epoch+1}/{num_epochs} | step {b_idx+1}/{len(train_loader)} "
                    f"| loss {avg_loss:.4f} | lr {lr:.3e}"
                )

        # flush any leftover grads if epoch ended mid-accumulation
        # (only needed if you want to force an update; often you do)
        if (global_step % config.grad_accum_steps) != 0:
            if config.max_grad_norm is not None:
                scaler.unscale_(optimizer)
                torch.nn.utils.clip_grad_norm_(model.parameters(), config.max_grad_norm)
            scaler.step(optimizer)
            scaler.update()
            optimizer.zero_grad(set_to_none=True)
            if scheduler is not None:
                scheduler.step()

        # epoch train metrics
        train_metrics = {k: v / max(1, epoch_steps) for k, v in epoch_sums.items()}
        train_metrics["epoch"] = epoch
        train_metrics["time_sec"] = time.time() - t0
        history["train"].append(train_metrics)

        # checkpoint: last
        if config.save_last:
            save_checkpoint(
                last_path,
                model=model,
                optimizer=optimizer,
                epoch=epoch,
                best_metric=best_metric,
                scaler=scaler,
                scheduler=scheduler,
            )

        # eval every N epochs (10 per your request), and also on the final epoch
        do_eval = (val_loader is not None) and (
            ((epoch + 1) % config.eval_every == 0) or ((epoch + 1) == num_epochs)
        )

        if do_eval:
            eval_metrics = evaluate(
                model=model,
                dataloader=val_loader,
                loss_fn=loss_fn,
                device=device,
                step_fn=step_fn,
                max_batches=max_eval_batches,
            )
            eval_metrics["epoch"] = epoch
            history["eval"].append(eval_metrics)

            monitor_value = float(eval_metrics.get(config.monitor, float("nan")))
            print(f"Eval @ epoch {epoch+1}: " + ", ".join(f"{k}={v:.4f}" for k, v in eval_metrics.items() if k != "epoch"))

            if config.save_best and (not torch.isnan(torch.tensor(monitor_value))):
                if is_better(monitor_value, best_metric):
                    best_metric = monitor_value
                    history["best_metric"] = best_metric
                    history["best_epoch"] = epoch
                    save_checkpoint(
                        best_path,
                        model=model,
                        optimizer=optimizer,
                        epoch=epoch,
                        best_metric=best_metric,
                        scaler=scaler,
                        scheduler=scheduler,
                        extra={"monitor": config.monitor, "mode": config.mode},
                    )
                    print(f"New best checkpoint saved: {best_path} ({config.monitor}={best_metric:.4f})")

    return history


# -----------------------------
# Usage example (you will adapt step_fn for HuBERT)
# -----------------------------
"""
config = TrainConfig(
    eval_every=10,
    grad_accum_steps=2,
    use_amp=True,
    ckpt_dir="./checkpoints",
    ckpt_name="hubert",
    monitor="loss",
    mode="min",
    resume_path=None,  # or "./checkpoints/hubert_last.pt"
)

history = train_eval_loop(
    model=model,
    train_loader=train_loader,
    val_loader=val_loader,
    num_epochs=10,
    loss_fn=loss_fn,
    optimizer=optimizer,
    scheduler=scheduler,   # optional
    config=config,
    step_fn=custom_hubert_step_fn,  # strongly recommended
)
"""
```
